```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read Excel file
df = pd.read_excel('TestCase1_Results.xlsx', sheet_name='Sheet1', index_col=0)

# Extract response times and configure settings
response_row = df.loc['Avg Response time']
scenarios = ['S1', 'S2', 'S3', 'S4', 'S5']
algorithms = ['ROUND-ROBIN', 'ESCE', 'Intra-balancer']
scenario_vms = [10, 25, 40, 50, 100]
bar_colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Build response_times matrix
response_times = []
for scenario in scenarios:
    scenario_times = [
        response_row[f"{scenario}_RR"],
        response_row[f"{scenario}_ESCE"],
        response_row[f"{scenario}_Intra"]
    ]
    response_times.append(scenario_times)

# Chart configuration
x = np.arange(len(scenario_vms))
bar_width = 0.18
gap = 0.04
max_response = max([max(scenario) for scenario in response_times])

# Create plot
fig, ax = plt.subplots(figsize=(14, 7))

# Plot data with precise spacing
for i, (algo, color) in enumerate(zip(algorithms, bar_colors)):
    offset = (i - 1) * (bar_width + gap)
    values = [rt[i] for rt in response_times]
    ax.bar(x + offset, values, bar_width, label=algo, color=color)

# Add scenario labels just above bars
for idx, (x_pos, scenario_data) in enumerate(zip(x, response_times)):
    max_in_group = max(scenario_data)
    label_y = max_in_group + (0.07 * max_response)  # 3% above tallest bar
    ax.text(
        x_pos, 
        label_y, 
        f'Scenario {idx+1}',
        ha='center', 
        va='bottom', 
        fontsize=10,
        fontweight='bold',
        fontfamily='sans-serif'
    )

# Customize chart
ax.set_xlabel('Number of Allocated VMs', fontsize=12, labelpad=10)
ax.set_ylabel('Average Response Time (ms)', fontsize=12, labelpad=10)
ax.set_title('Overall Average Response Time Comparison (Case 1)', fontsize=14, pad=20, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(scenario_vms)
ax.set_ylim(0, max_response * 1.12)  # Tight y-limit
ax.legend(loc='upper right', frameon=True)

# Add value labels
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', padding=2, fontsize=9)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------------------
# 1) DATA PREPARATION
# ---------------------------------------------------------------------
scenarios = ["Scenario 1", "Scenario 2", "Scenario 3", "Scenario 4", "Scenario 5"]
vm_counts = [10, 25, 40, 50, 100]  # X-axis tick labels

existing_response = np.array([140.98, 111.91, 105.54, 102.69, 98.93])
our_response      = np.array([140.27, 111.46, 104.87, 102.97, 98.03])

x_positions = np.arange(len(scenarios))

# ---------------------------------------------------------------------
# 2) FIGURE AND AXIS SETUP
# ---------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.title("Overall Average Response Time Comparison (Case 1)", fontsize=14)
plt.xlabel("Number of Allocated VMs", fontsize=12)
plt.ylabel("Average Response Time (ms)", fontsize=12)

# Replace default x-axis ticks with the VM counts
plt.xticks(x_positions, vm_counts)

plt.ylim(0, 160)

# ---------------------------------------------------------------------
# 3) PLOTTING THE GROUPED BARS
# ---------------------------------------------------------------------
bar_width = 0.8
offset = bar_width / 2

bars_existing = plt.bar(
    x_positions - offset/2, 
    existing_response, 
    width=bar_width/2, 
    color="purple", 
    label="Existing Implementation"
)
bars_our = plt.bar(
    x_positions + offset/2, 
    our_response, 
    width=bar_width/2, 
    color="orange", 
    label="Our Implementation"
)

# ---------------------------------------------------------------------
# 4) LABEL EACH BAR WITH ITS VALUE
# ---------------------------------------------------------------------
def label_bars(bars):
    """Attach a text label above each bar displaying its height."""
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + 2, 
            f"{height:.2f}",
            ha='center',
            va='bottom',
            fontsize=9,
            color='black'
        )

label_bars(bars_existing)
label_bars(bars_our)

# ---------------------------------------------------------------------
# 5) PLACE SCENARIO LABELS ABOVE EACH BAR GROUP
# ---------------------------------------------------------------------
for i, scenario in enumerate(scenarios):
    max_height = max(existing_response[i], our_response[i])
    plt.text(
        x_positions[i], 
        max_height + 10,  
        scenario,
        ha='center',
        va='bottom',
        fontweight='bold'
    )

# ---------------------------------------------------------------------
# 6) LEGEND, GRID, AND LAYOUT
# ---------------------------------------------------------------------
plt.legend(loc="upper right")
plt.grid(axis='y', linestyle='', alpha=0.5)

# Manually adjust the layout to avoid label cut-off issues
plt.subplots_adjust(top=0.9, bottom=0.15)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# ---------------------------------------------------------------------
# 1) DATA PREPARATION
# ---------------------------------------------------------------------
scenarios = ["Scenario 1", "Scenario 2", "Scenario 3", "Scenario 4", "Scenario 5"]
x_labels = [5, 10, 15, 20, 25]  # X-axis tick labels

existing_processing = np.array([2.27, 1.80, 1.73, 1.67, 1.69])
our_processing      = np.array([2.12, 1.76, 1.68, 1.60, 1.57])

x_positions = np.arange(len(scenarios))

# ---------------------------------------------------------------------
# 2) FIGURE AND AXIS SETUP
# ---------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.title("Overall Average Processing Time Comparison (Case 2)", fontsize=14)
plt.xlabel("Number of User Requests", fontsize=12)
plt.ylabel("Average Processing Time (ms)", fontsize=12)

plt.xticks(x_positions, x_labels)

plt.ylim(0, 3.0)

# ---------------------------------------------------------------------
# 3) PLOTTING THE GROUPED BARS
# ---------------------------------------------------------------------
bar_width = 0.8
offset = bar_width / 2

bars_existing = plt.bar(
    x_positions - offset/2, 
    existing_processing, 
    width=bar_width/2, 
    color="purple", 
    label="Existing Implementation"
)
bars_our = plt.bar(
    x_positions + offset/2, 
    our_processing, 
    width=bar_width/2, 
    color="orange", 
    label="Our Implementation"
)

# ---------------------------------------------------------------------
# 4) LABEL EACH BAR WITH ITS VALUE
# ---------------------------------------------------------------------
def label_bars(bars, offset=0.2):
    """
    Attach a text label above each bar displaying its height.
    'offset' is how far above the bar to place the label.
    """
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height + offset,   # Slightly above the bar
            f"{height:.2f}",
            ha='center',
            va='bottom',
            fontsize=9,
            color='black'
        )

label_bars(bars_existing, offset=0.04)
label_bars(bars_our, offset=0.04)

# ---------------------------------------------------------------------
# 5) PLACE SCENARIO LABELS ABOVE EACH BAR GROUP
# ---------------------------------------------------------------------
for i, scenario in enumerate(scenarios):
    max_height = max(existing_processing[i], our_processing[i])
    plt.text(
        x_positions[i], 
        max_height + 0.2,  
        scenario,
        ha='center',
        va='bottom',
        fontweight='bold'
    )

# ---------------------------------------------------------------------
# 6) LEGEND, GRID, AND LAYOUT
# ---------------------------------------------------------------------
plt.legend(loc="upper right")

plt.grid(axis='y', linestyle='', alpha=0.5)

plt.subplots_adjust(top=0.9, bottom=0.12)

plt.show()
```
